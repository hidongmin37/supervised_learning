# ARMA 모델
* array 형태로 작성해야

```python
import pandas as pd
import FinanceDataReader as fdr
import matplotlib.pyplot as plt
```

```python
df_kakao=fdr.DataReader('035720',start='2010')
```

```python
df_kakao['Close'].plot(figsize=(20,10))
plt.title('Close Price for Apple')
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
plot_acf(df_kakao['Close'])
plot_pacf(df_kakao['Close'])
plt.show()
```

```python
diff_1=df_kakao['Close'].diff(periods=1).iloc[1:]
diff_1.plot()
plot_acf(diff_1)
plot_pacf(diff_1)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model= ARIMA(df_kakao['Close'],order=(1,1,0)) # 0,1,1 # 1,1,0
model_fit = model.fit()
print(model_fit.summary())
```

```python
model= ARIMA(df_kakao['Close'],order=(0,1,1)) # 0,1,1 # 1,1,0
model_fit = model.fit()
print(model_fit.summary())
```

```python
model= ARIMA(df_kakao['Close'],order=(1,1,1)) # 0,1,1 # 1,1,0
model_fit = model.fit()
print(model_fit.summary())
```
