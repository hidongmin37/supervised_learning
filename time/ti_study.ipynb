```python
import pandas as pd
import FinanceDataReader as fdr
import matplotlib.pyplot as plt

```

```python
df_kakao=fdr.DataReader('035720',start='2010')
```

```python
df_kakao.head()
```

```python
df_kakao.shift(periods=1, freq='D').head() 
```

```python
df_kakao['Close'].plot(figsize=(12,5)).autoscale(axis='x', tight=True)
```

```python
df_kakao.rolling(window=7).mean() #7일 기준 이동평균을 구하는 방법
 
df_kakao['Close'].plot(figsize=(12,5)).autoscale(axis='x', tight=True)
df_kakao.rolling(window=30).mean()['Close'].plot()
 
df_kakao['Close: 3D Day Mean'] = df_kakao['Close'].rolling(window=30).mean() #새로운 컬럼을 만들어서 plot으로 한꺼번에 같이 그리기
 
df_kakao[['Close', 'Close: 3D Day Mean']].plot(figsize=(12,5)).autoscale(axis='x', tight=True)
```
