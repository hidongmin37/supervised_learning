```python
import numpy as np
import pandas as pd
```

```python
# Datetimeindex
dates= pd.date_range('2020-01-01',periods=48,freq='M') #  M month end frequency

#additive model: trend + cycle + seasonality + irregular factor
timestamp= np.arange(len(dates))
trend_factor=timestamp*1.1
cycle_factor= 10*np.sin(np.linspace(0,3.14*2,48)) #여기서는 0부터 3.14*2를 48구간으로 나눈거지
#linspace는 1차원의 배열만들기 0부터 #np.linspace(1,2,100) 1과 2사이를 숫자 100개
seasonal_factor=7*np.sin(np.linspace(0,3.14*8,48))
np.random.seed(2004) #고정된 값
irregular_factor=2*np.random.randn(len(dates)) #len(dates)=48 난수지정
```

```python
timestamp
```

```python
df=pd.DataFrame({'timeseries':trend_factor+ cycle_factor+ seasonal_factor+irregular_factor,'trend':trend_factor,
                 'cycle':cycle_factor,
                 'seasonal':seasonal_factor,
                 'irregular': irregular_factor},
                index=dates)
df
```

### 추세변동

```python
# Time series plot
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=[10,6])
df.plot()
plt.title('Time Series(Additive Model)',fontsize=16)
plt.ylim(-12,55)
plt.show()
```
