##### 불균형한 DataSet을 위한 KFold방법이다. 균형있게 나눈다고 생각할 수 있음
##### 교차 검증이라는 것 자체를 알고리즘을 다르게 해서 할 수 있음

```python
from sklearn.model_selection import StratifiedKFold
import numpy as np
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris=load_iris()
```

```python
import pandas as pd

iris_df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
iris_df
```

```python
iris_df['label']=iris.target #레이블 0,1,2로 나오는 것들 붙여 줌
iris_df['label'].value_counts()
```

```python
iris_df
```

```python
from sklearn.model_selection import StratifiedKFold 


skf= StratifiedKFold(n_splits=3) #똑같이 3개로 나눠서 정해줌
n_iter=0

for train_index,test_index in skf.split(iris_df,iris_df['label']):
    n_iter +=1
    label_train =iris_df['label'].iloc[train_index]
    label_test=iris_df['label'].iloc[test_index]
    print(f'#{n_iter} 교차 검증  {0}')
    print('학습 레이블 데이터 분포:\n', label_train.value_counts())
    print('검증 레이블 데이터 분포:\n', label_test.value_counts())
```

```python
####K폴드와 비교
from sklearn.model_selection import KFold
kfold = KFold(n_splits=3)
n_iter = 0
for train_index,test_index in kfold.split(iris_df):
           n_iter +=1
           label_train = iris_df['label'].iloc[train_index]
           label_test = iris_df['label'].iloc[test_index]
           print(f'#{n_iter} 교차 검증  {0}')
           print('학습 레이블 데이터 분포:\n', label_train.value_counts())
           print('검증 레이블 데이터 분포:\n', label_test.value_counts())
#케이폴드로 했을 때 학습이 더 잘 될 수도 있음
#케이폴드를 사용했을 때 데이터가 치우쳐져있음.
#데이터가 한쪽에 치우쳐져있어 satisfied폴드 검증 방법은 레이블 값을 비교하면서 
# 적절하게 섞으면서 해줌
```
