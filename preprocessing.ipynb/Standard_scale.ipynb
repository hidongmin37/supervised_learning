```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris= load_iris()
iris_data = iris.data
iris_df=pd.DataFrame(data=iris_data,columns=iris.feature_names)
#스캐터함수로변환
#스캐터 함수를 다시 변환
iris
```

```python
print('feature 들의 평균값')
print(iris_df.mean())
print(f'\n feature 들의 분산 값')
print(iris_df.var())
#데이터 자체를 때려박아도 괜찮음
```

```python
from sklearn.preprocessing import MinMaxScaler

iris_df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler() ###stanrdardScaler객체 생성
scaler.fit(iris_df)#iris_df로 훈련을 시킨다. fit은 데이터 변환을 위해서 기준 정보를 설정하는 역할

iris_scaled=scaler.transform(iris_df)  #transform()시 scale 변환된 데이터 셋이 numpy ndarray로 반환됨. 공식을 실행시켜준다고 생각하면됨.

iris_df_scaled =pd.DataFrame(data=iris_scaled,columns=iris.feature_names)#데이터 프레임화

iris_df

iris.feature_names
```

```python
print('feature 들의 평균 값')
print(iris_df_scaled.mean())
print('\n feature 들의 분산 값')
print(iris_df_scaled.var())
```
